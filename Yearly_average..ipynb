```python
# cleaning column names
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace('[^a-z0-9_ -]', '', regex=True)
df.columns = df.columns.str.replace('[ -]+', '_', regex=True)
df.columns = df.columns.str.replace('__+', '_', regex=True)
df.columns = df.columns.str.strip('_')
```

```python
# line plot of inflation spread through the years
plt.figure(figsize=(12, 6))
plt.plot(df['year'], df['12_month_inflation'], marker='o', linestyle='-')
plt.title('Inflation')
plt.xlabel('year')
plt.ylabel('12-Month Inflation (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# line plots to show trends over time
cols = ['12_month_inflation', 'total_debt', 'imports', 'exports',
        'foreign_direct_investment', 'unemployment_rate']

fig, axs = plt.subplots(3, 2, figsize=(18, 14))
axs = axs.flatten()

for i, col in enumerate(cols):
    if col in df.columns:
        sns.lineplot(data=df, x='year', y=col, ax=axs[i])
        axs[i].set_title(col.replace('_', ' ').title())
        axs[i].set_xlabel("Year")
        axs[i].set_ylabel("Value")
    else:
        print(f"Warning: Column '{col}' not found in DataFrame. Skipping plot.")


plt.tight_layout()
plt.show()
```

```python
sns.lineplot(data=df, x="year", y="exchange_rate", marker="o", color="blue")
plt.title("Exchange Rate")
plt.ylabel("Exchange Rate")
plt.xlabel("Year")
plt.tight_layout()
plt.show()
```

- It was noted that exchange rates have been rising over the years

```python
# a bar plot of total debt over the years
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="year", y="total_debt", color="steelblue")
plt.title("Total Debt")
plt.ylabel("Debt (USD)")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

-The country's total debt has been increasing over the years due to continued governments borrowing spree

```python
# scatter plot for exchange rates vs interest rates
sns.scatterplot(data=df, x="interest_rate", y="exchange_rate", hue="year", palette="coolwarm", s=100)
plt.title("Exchange Rate vs. Interest Rate")
plt.xlabel("Interest Rate (%)")
plt.ylabel("Exchange Rate")
plt.tight_layout()
plt.show()
```

```python
# Normalized indicators for comparison
normalized_df = df.copy()

columns_to_normalize = ["exchange_rate", "12_month_inflation", "interest_rate", "total_debt", "foreign_direct_investment"]
normalized_df[columns_to_normalize] = normalized_df[columns_to_normalize].apply(lambda x: (x - x.min()) / (x.max() - x.min()))

# Melted for multi-line plotting
df_melted = pd.melt(normalized_df, id_vars="year", value_vars=columns_to_normalize)

sns.lineplot(data=df_melted, x="year", y="value", hue="variable", marker="o")
plt.title("Normalized Trends of Key Economic Indicators")
plt.ylabel("Normalized Value")
plt.xlabel("Year")
plt.legend(title="Indicator")
plt.tight_layout()
plt.show()
```

```python
#pairplot multivariate relationship show relationship between multiple features
sns.pairplot(df[['12_month_inflation', 'total_debt', 'imports', 'exports', 'unemployment_rate']])
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```

```python
# ploting the correlation between the different columns
sns.pairplot(df)
```

```python
#distribution histograms check skewness
df.select_dtypes(include='number').hist(bins=15, figsize=(15, 10))
plt.suptitle("Distribution of Numeric Variables")
plt.show()
```

```python
# Correlation Heatmap
corr = df.corr()
fig, ax = plt.subplots(figsize=(14, 12))
cax = ax.imshow(corr, cmap='coolwarm', interpolation='nearest', vmin=-1, vmax=1)
cbar = fig.colorbar(cax, ax=ax, shrink=0.8)
cbar.set_label("Correlation Coefficient", fontsize=12)

# Set tick labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=45, ha='right', fontsize=10)
ax.set_yticklabels(corr.columns, fontsize=10)
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        value = corr.iloc[i, j]
        ax.text(j, i, f"{value:.2f}", ha='center', va='center', color='black')
ax.set_title("Correlation Heatmap", fontsize=16, weight='bold', pad=20)
plt.tight_layout()
plt.show()
```
