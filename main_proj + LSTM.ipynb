### Converting the columns to numeric

```python
#We shall make the columns numeric except for Date which we shall convert to datetime and make an index prior to beginnning visualization.

for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    for col in df.columns:
        if col != 'Date':
            df[col] = pd.to_numeric(df[col], errors='coerce')
```

### Handling missing values



```python
#We shall replace the missing data by mean, as the deviations are and should be very low across all the dataframes
for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            if df[col].isnull().any():
                df[col].fillna(df[col].mean(), inplace=True)
```

```python
#confirming it worked.
dfmonthly.isna().sum()
```

### Converting the Date Column to datetime and setting it as an index

```python
#We shall convert the Date columns in all the datasets to allow for visualization.
for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df.set_index('Date', inplace=True)
```

We shall combine the above plots as they are very closely inter-related.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=dfmonthly, x="Date", y="Savings", color="blue",label="Savings")
sns.lineplot(data=dfmonthly, x="Date", y="Deposit", color="black",label="Deposit")
sns.lineplot(data=dfmonthly, x="Date", y="Lending ", color="green",label="Lending")
sns.lineplot(data=dfmonthly, x="Date", y="Overdraft ", color="red",label="Overdraft")
plt.title("Savings, Deposit, Lending, Overdraft over Time")
plt.ylabel("Interest Rate")
plt.xlabel("Year")
plt.legend(title="Indicator")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#we shall stack the plots ontop of each other to get a clearer side by side analysis on the movements.
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 7), sharex=True)

sns.lineplot(data=dfmonthly, x="Date", y="Savings", color="blue", label="Savings", ax=ax1)
sns.lineplot(data=dfmonthly, x="Date", y="Deposit", color="black", label="Deposit", ax=ax1)
sns.lineplot(data=dfmonthly, x="Date", y="Lending ", color="green", label="Lending", ax=ax1)
sns.lineplot(data=dfmonthly, x="Date", y="Overdraft ", color="red", label="Overdraft", ax=ax1)

ax1.set_title("Savings, Deposit, Lending, Overdraft over Time")
ax1.set_ylabel("Interest Rate")
ax1.legend(title="Indicator")
ax1.grid(True)

sns.lineplot(data=dfmonthly, x="Date", y="exchange_rate", color="purple", ax=ax2)
ax2.set_title("Exchange Rate Over Time")
ax2.set_ylabel("Exchange Rate")
ax2.set_xlabel("Year")
ax2.grid(True)

plt.tight_layout()
plt.show()
```

There seems to be a pattern in the movement and as such it would be best to overlay both plots to see the movements much clearer.

```python
#overlaying the bank interest rates and the exchage rate
sns.set_style("whitegrid")

fig, ax1 = plt.subplots(figsize=(14, 6))

sns.lineplot(data=dfmonthly, x="Date", y="Savings", ax=ax1, label="Savings", color="blue", linewidth=1)
sns.lineplot(data=dfmonthly, x="Date", y="Deposit", ax=ax1, label="Deposit", color="black", linewidth=1)
sns.lineplot(data=dfmonthly, x="Date", y="Lending ", ax=ax1, label="Lending", color="green", linewidth=1)
sns.lineplot(data=dfmonthly, x="Date", y="Overdraft ", ax=ax1, label="Overdraft", color="red", linewidth=1)

ax1.set_ylabel("Interest Rate", fontsize=12)
ax1.set_xlabel("Year", fontsize=12)
ax1.set_title("Bank Specific Interest Rates and Exchange Rate Over Time", fontsize=14, fontweight="bold")

ax2 = ax1.twinx()
sns.lineplot(data=dfmonthly, x="Date", y="exchange_rate", ax=ax2,
             label="Exchange Rate", color="purple", linewidth=2, linestyle="--", alpha=0.8)

ax2.set_ylabel("Exchange Rate", fontsize=12)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(lines1 + lines2, labels1 + labels2, title="Indicator", loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=3)

ax1.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.6)
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)

plt.show()
```

```python
# Let's compute a correlation matrix.
cols = ["Savings", "Deposit", "Lending ", "Overdraft ", "exchange_rate"]
corr_matrix = dfmonthly[cols].corr()
print(corr_matrix)
```

Key Takeaways;

1. The deposit interast rate tends to follow closely the exchange rate, we can attribute this to the fact that money is weakening, less people are investing and as such due to the low demand for fixed deposits, banks are forced to increase the interest rate to attract more money.

2. Lending and Overdraft have very little effect on the exchange rate.

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.metrics import mean_squared_error, mean_absolute_error

# We will use the monthly data for the LSTM model
data = dfmonthly[['exchange_rate']].values

# Scaled the data to be between 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Created the training and testing data sets
training_data_len = int(np.ceil( len(scaled_data) * .80 ))

train_data = scaled_data[0:training_data_len, :]
look_back = 60
test_data = scaled_data[training_data_len - look_back: , :]

def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset) - look_back - 1):
        a = dataset[i:(i + look_back), 0]
        X.append(a)
        Y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(Y)

X_train, y_train = create_dataset(train_data, look_back)
X_test, y_test = create_dataset(test_data, look_back)

# Reshaped the data for the LSTM model
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Building the LSTM model

model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(25))
model.add(Dense(1))

# Compiled the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Trained the model
model.fit(X_train, y_train, batch_size=1, epochs=10)

# made predictions
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)

y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
# Evaluated the model
from sklearn.metrics import mean_squared_error, mean_absolute_error
rmse = np.sqrt(mean_squared_error(y_test_actual, predictions))
mae = mean_absolute_error(y_test_actual, predictions)
mse = mean_squared_error(y_test_actual, predictions)
print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
```

```python
# plots for exchange rate prediction
trainPredictPlot = np.empty_like(scaled_data)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(y_train)+look_back, :] = scaler.inverse_transform(model.predict(X_train))

testPredictPlot = np.empty_like(scaled_data)
testPredictPlot[:, :] = np.nan
test_predict_start_index = training_data_len + look_back - look_back
test_plot_start_index = training_data_len

testPredictPlot[test_plot_start_index:test_plot_start_index + len(predictions), :] = predictions

plt.figure(figsize=(14, 7))
plt.plot(dfmonthly.index, scaler.inverse_transform(scaled_data), label='Original Data')
plt.plot(dfmonthly.index, trainPredictPlot, label='Training Predictions')
plt.plot(dfmonthly.index, testPredictPlot, label='Testing Predictions')
plt.title('Exchange Rate Prediction using LSTM')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.show()

# a plot of actual vs predicted exchange rate
plt.figure(figsize=(14, 7))
test_indices = dfmonthly.index[training_data_len:training_data_len + len(y_test_actual)]
plt.plot(test_indices, y_test_actual, label='Actual Exchange Rate')
plt.plot(test_indices, predictions, label='Predicted Exchange Rate')
plt.title('Actual vs Predicted Exchange Rate (Test Set)')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# a forecast for exchange rates

def predict_future(model, last_sequence, num_future_steps, scaler):
    future_predictions = []
    current_sequence = last_sequence.copy()

    for _ in range(num_future_steps):
        # Reshaped for prediction
        predicted_value = model.predict(current_sequence.reshape(1, look_back, 1))

        # Inverse transformed the prediction
        predicted_value_scaled = scaler.inverse_transform(predicted_value)
        future_predictions.append(predicted_value_scaled[0, 0])

        # removed the oldest value, added the new prediction
        current_sequence = np.append(current_sequence[:, 1:, :], predicted_value.reshape(1, 1, 1), axis=1)

    return future_predictions

# Number of future steps to predict
num_future_months = 24

# Got the last 'look_back' data points from the scaled data
last_sequence = scaled_data[-look_back:].reshape(1, look_back, 1)

# Predicted the future
future_forecast = predict_future(model, last_sequence, num_future_months, scaler)

# Created a date range for the future predictions
last_date = dfmonthly.index[-1]
future_dates = pd.date_range(start=last_date, periods=num_future_months + 1, freq='M')[1:]

# Created a DataFrame for the future forecast
future_forecast_df = pd.DataFrame({'exchange_rate': future_forecast}, index=future_dates)

```

```python
# predicted future values with multi-input features
def predict_future_multi(model, last_sequence, num_future_steps, scaler, features):
    future_predictions = []
    current_sequence = last_sequence.copy()

    # Got the index of the target column
    target_idx = features.index('exchange_rate')

    for _ in range(num_future_steps):
        # Predicted the next step for the target feature
        predicted_value_scaled_target = model.predict(current_sequence)

        # Inversed and  transformed the predicted target value

        dummy_array_pred = np.zeros((1, len(features)))
        dummy_array_pred[0, target_idx] = predicted_value_scaled_target[0, 0]
        predicted_value_actual_target = scaler.inverse_transform(dummy_array_pred)[0, target_idx]

        future_predictions.append(predicted_value_actual_target)


        # used the last known values of other features from the input sequence for the new step.
        new_step_scaled = np.copy(current_sequence[0, -1, :])
        new_step_scaled[target_idx] = predicted_value_scaled_target[0, 0]
        new_step_scaled = new_step_scaled.reshape(1, 1, -1)


        # removed the oldest timestep, added the new timestep
        current_sequence = np.append(current_sequence[:, 1:, :], new_step_scaled, axis=1)


    return future_predictions

# Number of future steps to predict
num_future_months_multi = 24

# Got the last 'look_back' data points from the scaled data with all features
last_sequence_multi = scaled_data_multi[-look_back:].reshape(1, look_back, scaled_data_multi.shape[1])

# Predicted the future
future_forecast_multi = predict_future_multi(model_multi, last_sequence_multi, num_future_months_multi, scaler_multi, features)

# Created a date range for the future predictions
last_date_multi = dfmonthly.index[-1]
future_dates_multi = pd.date_range(start=last_date_multi, periods=num_future_months_multi + 1, freq='M')[1:]

# Created a DataFrame for the future forecast
future_forecast_df_multi = pd.DataFrame({'exchange_rate': future_forecast_multi}, index=future_dates_multi)

```
